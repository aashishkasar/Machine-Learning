### Rotation

```python
img=cv2.imread(r"D:\Data Sets\images\timon.jpg")
```

```python
img.shape[0]/2
```

```python
img.shape[1]/2
```

```python
rotation=cv2.getRotationMatrix2D((img.shape[0]/2,img.shape[1]/2),180,0.5)
```

Choose middle value for center value, pass center of image, second pass rotation angle, 1 for same scale â˜

```python
imgt=cv2.warpAffine(img,rotation,img.shape[0:2])  #-->img.shape--for chane shape
cv2.imshow("s",imgt)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
