```python
import re
import emoji
from textblob import TextBlob
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer,WordNetLemmatizer
```

#### 1.Here we are using `regex` for removing unnecessary characters in text 
#### 2. `emoji` using this we are decoding the information from emoji present in the text
#### 3. `TextBlob` using textblob we are correcting the spelling of english words
#### 4. `Stem` using stemming we are storing the root word of a particular word 
#### 5. `tokenize` using this dividing paragraph or sentences  into tokens

```python
data = data[data['Language']=='en']
```

#### convert to english

```python
data
```

```python
data.isnull().sum()
```

```python
stop_words = stopwords.words("english")
stop_words.remove('not')
```

#### removing `not` in the stopwords 

```python
%%time
sentances = []
for i in data['Text']:
    word = []
    sent = i.lower()                           #first converting into lower case
    sent = emoji.demojize(sent)             #getting desired information from the emojis
    sent = re.sub('â€˜|â€™',"'",sent)         
    sent = re.sub('&amp;',"and",sent)
    sent = re.sub('<.*?>','',sent)                # removing unnecessary characters
    sent = re.sub('https?://\S*|www\S*','',sent)
    sent = re.sub(' @\w+| #\w+',"",sent)
    sent = re.sub("[^a-z0-9 ]","",sent)
    for i in word_tokenize(sent):                   #using word_tokenize dividing into words
        if i in stop_words:
            pass
        else:word.append(PorterStemmer().stem(i))  #neglecting the words which are in the stopwords
    sentances.append(" ".join(word))                 #Appending into the list
```

```python
cleaned_text = pd.DataFrame({"cleaned_Text":sentances})
```

```python
cleaned_text
```

#### bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

#### using `CountVectorizer` converting text into vectors

```python
count_obj=CountVectorizer()
trans=count_obj.fit_transform(cleaned_text['cleaned_Text'][0:25000])
trans
```

```python
%%time
array_count = trans.toarray()
```

```python
count_obj.vocabulary_
```

```python
pd.DataFrame(array_count,columns=sorted(count_obj.vocabulary_))
```

#### TF-IDF

```python
tf= TfidfVectorizer()
trans_tf = tf.fit_transform(cleaned_text['cleaned_Text'][0:5000])
trans_tf
```

```python
pd.DataFrame(trans_tf.toarray(),columns=sorted(tf.vocabulary_.keys()))
```
